## Fundamentals of Deep Learning

## What are neural networks?

Neural networks are a set of algorithms, modeled loosely after the human brain, that are designed to recognize patterns. They take in raw data as input, form multiple weighted combinations and transformations of this data before arriving at the right values of parameters that can achieve a specific outcome. 

Neural neworks are typically organized in layers of neurons, where, a neuron is a cell that:

1. Receives inputs from the previous layer
2. Computes a weighted combination (sum) of inputs and
3. Applies a simple transformation function (called activation function) 

Then, passes the result on to the next layer.  

Data is presented to the network via the 'input layer'. Then comes, one or more 'hidden layers' where the actual 'learning' of parameters happen. 

The hidden layers then links to a final 'output layer' from where we finally obtain the result.

Below is an image of a typical Artificial Neural Network, aka, ANN. This is the fundamental design. From this basic structure, modifications in design can be made to form numerous different architectures.

## What is deep learning?

Deep learning is accomplished *using* neural networks. It is a machine learning framework that utilizes multiple layers to extract features from raw input data.

It is a branch of artificial intelligence that implements 'learning by example', which is very similar to how human beings learn as well. This is why deep learning makes it possible to input raw data into a computer and get conclusive results on this data without any human interference.

The most popular examples of deep learning are voice recognition softwares (think Siri and Alexa), driverless cars and facial recognition software.

## Neurons

Neurons are the building blocks of a neural network. They taken in raw input data, pass them through a mathematical transformation and pass on the output to the next layer in neural  network.

The neurons present in every layer are interconnected, each's input characterised by a specific weight and bias. The image below shows a representation of a neuron, which takes in inputs and gives an output after mathematical transformation.

```python
Image('neuron2.png')
```

Each neuron is simple the combination of two mathematical functions, namely, the affine function and the activation function. 
The input data first goes through an affine transformation, and the output of affine function is the input to activation function. 
The final output from the neuron is the output of activation function.

The affine transformation is simply the weighted sum of inputs with the bias added.

As you can see, the first step is the affine transformation. The output after affine transformation is the input for activation function. 
There can be different activation functions.
The most popular ones include ReLU, sigmoid and tanh.

## What is PyTorch?

PyTorch is a deep learning framework, similar to Tensorflow. It was developed by Facebook in 2016 and since then has been steadily gaining popularity as an alternaative to Tensorflow.

Tensorflow is the earliest deep learning framework invented, and it was developed by Google. Even today it is the first library people learn about when entering the field of deep learning. Still, PyTorch is preferred by many machine learning developers for the following reasons:

1. PyTorch is *pythonic*. This means that the PyTorch library is built in close integration with the python language and python developers will find it easier to work with PyTorch for developing deep learning models. You can also use python's existing libraries like NumPy and SciPy to further enhance PyTorch's functionalities.


2. PyTorch supports *dynamic computation graphs*. While static computational graphs (like those used in TensorFlow) are defined prior to runtime, dynamic graphs are defined "on the fly" via the forward computation. In other words, the graph is rebuilt from scratch on every iteration . Dont worry if you dont understand much about it at this stage, you will read more about it later in the tutorial.


3. PyTorch is *easy to learn* because of its simple and intuitive syntax and close resemblance to the python language.


4. PyTorch allows for *easy debugging* because most of the debugging tools in python can also be used in PyTorch.

## Comparison between PyTorch and Tensorflow

**1. Computation Graph**

This is by far the most important difference between TensorFlow and PyTorch. 

TensorFlow uses static computation graph while PyTorch uses dynamic computation graph. What this means in the easiet terms is that while the parameters of computation graph (which is the graph formed after interconnections between all nodes in all layers of a neural network) can be changed during the runtime in PyTorch, the same is not possible in TensorFlow.

Thus PyTorch is said to have a *dynamic* computation graph.

**2. Debugging**

Debugging is easier in PyTorch as compared to TensorFlow. This is because PyTorch allows the user to make use of the standard debugging tools in python, such as pdb, ipdb, PyCharm debugger or old trusty print statements.

In contrast, in TensorFlow you have to use a special tool called tfdbg which allows to evaluate tensorflow expressions at runtime and browse all tensors and operations in session scope. But, you wont be able to debug any python code with it, so it will be necessary to use pdb separately.

**3. Visualization**

Tensorflow has its own visualisation framework known as TensorBoard. On the other hand, PyTorch makes use of the standard visualisation libraries such as Seaborn and Matplotlib.

**4. Data Parallelism**

It is much easier to achieve data parallelism in PyTorch than in Tensorflow. You can use `torch.nn.DataParallel` to wrap any module and it will be parallelized over batch dimension. This way you can leverage multiple GPUs with almost no effort.

On the other hand, TensorFlow requires you to fine tune every operation to be run on specific device. Thus, defining parallelism is way more manual and requires careful thought.

## PyTorch support for CUDA

For understanding what is CUDA support, you first need to understand the meaning and significance of GPU.

**GPU** stands for graphical processing unit. These are specialised chips with highly paralle architecture that makes them and order of magnitude faster than CPUs (central processing unit) for deep learning applications.
PyTorch Tensors have been architected to make optimal use of GPUs for massively parallel computations.

GPUs can speed up deep learning operations by almost 10-50X using parallel computing.

Now coming back to CUDA, **CUDA** has been devised by Nvidia, and is a parallel computing platform and API.
Nvidia is the leading manufacturer of GPUs across the globe.

How CUDA support works is:
* Developers can write CUDA-compliant code. 
* Code must be understood by CUDA aware framework (e.g. PyTorch). 
* If CUDA-enabled GPUs are available, speedup will automatically occur.

To check whether your device is CUDA compatible, you can run `torch.cuda.is_available()`. If it returns True, this means you can allocate your GPU for deep learning applications.

In case you dont have a GPU on your local computer, you can always use a cloud platform (offered by Google and Amazon) to run your model using a virtual GPU.

## Autograd in PyTorch

For finding the weights and biases of each input variable in the neural network, we use gradient descent optimization.

The aim of gradient descent optimization is to reduce the MSE (Mean Squared Error) for the neural network model developed. For this, we use forward and backward passes.


```python
Image('graddes.png')
```

## Forward Passes

In the forward pass, the current weights and biases are used to make a prediction using the neural network. For the first iteration weights and biases are usually taken to be 0.

Once we get the y_pred (predicted output), we calculate the error in comaprison to y (actual value).
Gradient Descent optimization makes use of these errors to tweak the parameters and reduce the error in consequent iterations.

## Backward Passes

A backward pass is when the error from forward pass is used to update parameter values for each layer. 
A backward pass starts from the last layer and goes on to the second last layer, third last layer and so on.

```python
Image('forbackpasses.png')
```

The backward pass allows the weights and biases of the neurons to converge to their final values.

To perform this whole process in PyTorch, we use the Autograd library.

## Autograd

The mathematical expression used during gradient descent optimization is shown below.

```python
Image('autograd.png')
```

Here, the gradient can be calculated in 3 ways.

* Numeric Differentiation
* Symbolic Differentiation
* Automatic Differentiation

AutoGrad in PyTroch uses automatic differentiation, as this process of calculating a gradient is easiest to implement, even though mathematically complex.

It makes use of Taylor's series expansion that allows fast approximation of parameters. The actual math can be very complex so its enough to know that auto-differentiation makes use of 2 passes, one forward to calculate and optimize the error, and one backward to update parameters.

Back propagation is only required during training: to do so in PyTorch, invoke the `.backward()` method.

Another thing worth noting is that
`requires_grad` is an attribute of the Variable class. By default, its False. 

It comes handy when you have to freeze some layers, and stop them from updating parameters while training. You can simply set the requires_grad to False, and these Variables wont be included in the computation graph. Thus, no gradient would be propagated to them, or to those layers which depend upon these layers for gradient flow. 

`requires_grad`, when set to True is contagious, meaning even if one operand of an operation has requires_grad set to True, so will the result.

```python
tensor1 = torch.Tensor([[1, 2, 3], 
                       [4, 5, 6]])
tensor1
```

```python
tensor2 = torch.Tensor([[7, 8, 9], 
                        [10, 11, 12]])

tensor2
```

By default, `.requires_grad` attribute is false.

```python
tensor1.requires_grad
```

```python
tensor2.requires_grad
```

We can set this attribute to True by calling the `_requires_grad()` function.

```python
tensor1.requires_grad_()
```

```python
tensor2.requires_grad_()
```

The `.grad` property stores all the gradients for the tensor.

```python
print(tensor1.grad)
```

As you can see, there are no gradients yet.

The `.grad_fn` property contains the gradient function.

This is not set either.

```python
print(tensor1.grad_fn)
```

If you create a new tensor using some mathematical operation, you will notice that there will be a gradient function, but still no gradients.

```python
output_tensor = tensor1 * tensor2
```

```python
print(output_tensor.grad_fn)
```

```python
print(output_tensor.grad)
```

Changing the operation for the output changes the gradient function.

The gradient function only contains the last operation. Here, even though there is a multiplication as well as a mean, only the mean calculation is recorded as the gradient function.


```python
output_tensor = (tensor1 * tensor2).mean()
print(output_tensor.grad_fn)
```

To calculate the gradients, we need to explicitly perform a backward propagation.

```python
output_tensor.backward()
```

The gradients are now available for the input tensor.
Future calls to backward will accumulate gradients into this vector.


```python
print(tensor1.grad)
```

The gradient vector is the same shape as the original vector.

```python
tensor1.grad.shape, tensor1.shape
```
